```python
import pandas as pd # For data manipulation and analysis
pd.set_option('display.max_columns', 500)
import numpy as np # For scientific computing
```

```python
ind = pd.read_csv(r'C:\Users\kwonk\Guided Project\KNIME_Project\Round 3\Data\Two-Digit Industries\2004_Main_Industry_Performance.csv')
ind = ind.dropna(subset=['Region'])
ind.fillna(0,inplace=True)
ind
```

```python
def gini_calc(column):
    df = ind[column]
    df = df/df.sum()
    sums_total=0
    
    constant=1 / (2 * (31*31) * df.mean() )
    
    for index1 in df.index:
        for index2 in df.index:
            sums=abs(df.loc[index1]-df.loc[index2])
            sums_total=sums_total+sums
    gini=sums_total*constant
    return gini
```

```python
# gini,df=gini_calc('Output (Fur, Feather, Leather)')

gini_calc(ind.columns[1])
```

```python
output=pd.DataFrame()

for column in ind.columns:
    if column != 'Region':
        gini = gini_calc(column)
        print(column)
        print(gini)
```
